# Lecture5(Filter rows by column value)

```python
imdb=pd.read_csv('http://bit.ly/imdbratings')
```

```python
imdb.head()
```

Find the lowest rated movied with genre Action and content rating R having duration more than 120

```python
###Using for loop
checklist=[]
for i in range(imdb.shape[0]):
    if imdb.content_rating[i]=='R' and imdb.genre[i]=='Action' and imdb.duration[i]>120:
        checklist.append(True)
    else:
        checklist.append(False)
imdb[checklist].sort_values('star_rating').head()
```

```python
###without using for loop 
duration_imdb=imdb[imdb.duration>120]
rating_imdb=duration_imdb[duration_imdb.content_rating == 'R']
genre_imdb=rating_imdb[rating_imdb.genre == 'Action']
genre_imdb.sort_values('star_rating')
```
