```python
import pandas as pd
import matplotlib as mt
import numpy as np
```

```python
data=pd.read_csv('Chicago_Crime_Detective.csv')
```

```python
data['Date'] =pd.to_datetime(data.Date)
data=data.sort_values(by="Date")
mid=int((data.shape[0]+1)/2)
data.iloc[mid-1:mid+2,:]
#median month, year->May, 2006
```

```python
data['month'] = pd.DatetimeIndex(data['Date']).month
data['month'].value_counts()
#lowest month is ->February
```

```python
# data.drop(columns=['weekday'])
# data.dtypes
```

```python
data['weekday'] = data['Date'].dt.dayofweek
data
```

```python
print(data['weekday'].value_counts())
data['Year'].value_counts()
#Most theft occured on Friday(4)
```

```python
arrest_done=data[data.Arrest==True]
arrest_done['month'].value_counts()
#for month of january thepts most arrests made
```
